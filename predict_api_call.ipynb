```python
import requests
import pandas as pd

input_text = '''
Am Tue, 16. 05. 22 10:53:31 +0100 schrieb Herbert Auer:


"Fußball ist wie Schach, nur ohne Würfel" (Theodor Fähling)

Größe,

Herbert
'''

url = "http://pseugc-app.project.ris.bht-berlin.de/predict"
payload = {
    "input_texts": [input_text],
    "repeat": 1
}
response = requests.post(url, json=payload)

# print("Status Code:", response.status_code)
# print("Response Body:", response.json())


display(pd.DataFrame(response.json()['output'][0][0]['output_dict']))
print(response.json()['output'][0][0]['output_text'])
```
